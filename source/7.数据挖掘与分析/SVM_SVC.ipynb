```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis');
plt.scatter(1.0,4.0,c="red",marker='x',s=60)
```

```python
from sklearn import svm
clf = svm.SVC(gamma='auto')
clf.fit(X,y)

print(clf.predict([[1.0,4.0]]))
```

```python
print(clf.n_support_)
print(clf.support_)
print(clf.support_vectors_)
```

```python
import numpy as np

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis');
plt.scatter(1.0,4.0,c="red",marker='x',s=60)
print(clf.n_support_[0])
X1=X[clf.support_[0:clf.n_support_[0]]]
X2=X[clf.support_[clf.n_support_[0]:]]

#改变支持向量的顺序，使其围成一个多边形
#如果数据集变了，需要根据支持向量的特征调整
X1[[0,3],:]=X1[[3,0],:];X1[[3,4],:]=X1[[4,3],:];X1=np.vstack((X1,X1[0]))
X2[[1,3], :] = X2[[3,1], :];X2=np.vstack((X2,X2[0]))

plt.plot(X1[:,0],X1[:,1],'--')
plt.plot(X2[:,0],X2[:,1],'--')
```
