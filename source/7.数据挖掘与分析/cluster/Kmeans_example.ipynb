```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
data, target = make_blobs(n_samples=80, n_features=2, centers=5, cluster_std=0.7) 
plt.scatter(data[:,0], data[:,1])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(data)
y_types = kmeans.predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=y_types, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='green', s=200, alpha=0.5)
```

```python
labels = KMeans(7, random_state=0).fit_predict(data)
plt.scatter(data[:, 0], data[:, 1], c=labels, s=50, cmap='viridis');
```

```python
from sklearn.cluster import KMeans

def kcluster(k,data):
    model = KMeans(n_clusters=k)
    model.fit(data)
    md = model.inertia_ / data.shape[0]
    return md

import matplotlib.pyplot as plt

x=range(2,10)
y=[]
for i in x:
    y.append(kcluster(i,data))
    
plt.plot(x,y)
for i,j in zip(x,y):
 plt.text(i+0.4, j+0.05, '%.3f' % j, ha='center', va= 'bottom',fontsize=11)
plt.axis([2,10,0,10])
plt.show()
    

```
