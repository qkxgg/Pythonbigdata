```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=100, centers=3,random_state=0, cluster_std=1.05)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis');
```

```python
#选取参与训练和预测的特征与目标
features = X
target = y
#分离数据集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=33)
```

```python
#从sklearn中引入支持向量机分类（SVC）
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))

import numpy as np


x_result=np.concatenate((x_train,x_test),axis=0)
y_result=np.concatenate((y_train,y_pred),axis=0)
plt.scatter(x_result[:, 0], x_result[:, 1], c=y_result, s=50, cmap='viridis');
plt.plot(svc.support_vectors_[:,0],svc.support_vectors_[:,1])
```

```python
print(svc.support_vectors_)
print(svc.support_)
print(svc.n_support_)
```

```python
#从sklearn中引入支持向量机分类（SVC）
from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#从sklearn中引入支持向量机分类（SVC）
from sklearn.svm import SVC
svc = SVC(kernel='poly')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
