```python
from sklearn import linear_model
X=[[0, 0], [1, 1], [2, 2]]
y=[0.5, 1.3, 2.4]
reg = linear_model.LinearRegression()

model= reg.fit(X,y)

print(model.coef_,model.intercept_)
print(model.predict([[1,3]]))
```
